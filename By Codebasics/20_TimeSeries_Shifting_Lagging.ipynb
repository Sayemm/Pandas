```python
import pandas as pd
```

```python
df = pd.read_csv("fb.csv", parse_dates=['Date'], index_col = 'Date')

df
```

```python
df.shift(3)
```

```python
df.shift(-1)
```

```python
df['prev'] = df['Price'].shift(1)

df
```

```python
df['1 day'] = df['Price']-df['prev']

df
```

```python
df['5 day % return'] = (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)

df
```

```python
df = df[['Price']]
df
```

```python
df.index
```

```python
df.index = pd.date_range(start = '2017-08-15', periods = 10, freq = 'B')

df.index
```
