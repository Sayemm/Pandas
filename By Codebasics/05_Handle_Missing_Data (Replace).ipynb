```python
import pandas as pd
import numpy as np

df = pd.read_csv("weather_data_replace.csv")

df
```

```python
# new_df = df.replace([-99999, -8888, '0'], np.NaN)

new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : '-99999',
    'event' : '0'
}, np.NaN)

new_df
```
